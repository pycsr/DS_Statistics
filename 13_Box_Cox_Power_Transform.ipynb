```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import probplot,boxcox
```

# Generate Random data which is not following Gausian Dirstribution

```python
a = np.random.exponential(size = 5000)
```

## kde plot

```python
sns.kdeplot(a)
```

## Q-Q Plot

```python
probplot(a,plot=plt)
plt.show()
```

## Apply Box-Cox transform to transform non-Gaursian data to Gausian Normal Distribution

```python
tr,lam = boxcox(a)
```

## kde plot

```python
sns.kdeplot(tr)
```

## Q-Q plot

```python
probplot(tr,plot=plt)
plt.show()
```
