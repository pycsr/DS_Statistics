# Q-Q Plot
- Statistics, Q-Q(quantile-quantile) plots play a very vital role to graphically analyze and compare two probability distributions by plotting their quantiles against each other. If the two distributions which we are comparing are exactly equal then the points on the Q-Q plot will perfectly lie on a straight line y = x

- You can tell the type of distribution using the power of the Q-Q plot just by looking at the plot
- You can tell the mean of the Random data set plot just by looking at the plot, by referring mean from Quantile plot of Gaussian Normal distribution


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import probplot
```

```python
a = np.random.normal(loc=0,scale=1,size=500)
# a
```

```python
sns.kdeplot(a)
```

```python
probplot(a,plot=plt)
plt.show()
```

## lets add more samples

```python
a = np.random.normal(loc=0,scale=1,size=5000)
probplot(a,plot=plt)
plt.show()
```

```python
a = np.random.normal(loc=0,scale=1,size=50000)
probplot(a,plot=plt)
plt.show()
```

## lets change the mean and std of Normal distribution

```python
a = np.random.normal(loc=10,scale=5,size=5000)
probplot(a,plot=plt)
plt.show()
```

```python
a = np.random.normal(loc=10,scale=5,size=50000)
probplot(a,plot=plt)
plt.show()
```

## Take Random data which is not from normal distribution

```python
a = np.random.randint(1,200,size=500)
probplot(a,dist="norm",plot=plt)
plt.show()
```
