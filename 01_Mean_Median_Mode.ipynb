# Measure of Central tendency

## Mean

```python
a = [1,2,3,4,5,6]
mean(a)
```

```python
a = np.arange(1,7)
a
```

```python
a.mean()
```

```python
sns.kdeplot(a)
```

```python
sns.kdeplot(a)
plt.vlines(x=a.mean(),ymin=0,ymax=.16,color="r")
```
