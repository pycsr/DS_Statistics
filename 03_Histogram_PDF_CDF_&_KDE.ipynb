```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Histogram

```python
df = sns.load_dataset("iris")
df.head(3)
```

```python
sns.histplot(df.petal_length)
```

```python
sns.histplot(df.petal_length,bins=5)
```

# Probability Density Function (PDF)

```python
count,bin_size = np.histogram(df.petal_length)
print(count)
print(bin_size)
```

```python
pdf = count/sum(count)
pdf
```

```python
plt.plot(pdf)
plt.xticks(range(10))
plt.show()
```

# Cummulative Density Function (CDF)

```python
count,bin_size = np.histogram(df.petal_length)
print(count)
print(bin_size)
```

```python
pdf = count/sum(count)
pdf
```

```python
cdf = np.cumsum(pdf)
cdf
```

```python
plt.plot(pdf)
plt.plot(cdf)
```

```python
count,bin_size = np.histogram(df.petal_length[df.species=="setosa"])
print(count)
print(bin_size)
```

```python
pdf = count/sum(count)
pdf
```

```python
cdf = np.cumsum(pdf)
cdf
```

```python
plt.plot(pdf)
plt.plot(cdf)
```

```python
sns.displot(x=df.petal_length,hue=df.species,kind="kde",rug=True)
```
