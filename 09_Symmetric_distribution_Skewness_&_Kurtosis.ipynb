# Skewness
- In statistics, skewness is a degree of asymmetry observed in a probability distribution that deviates from the symmetrical normal distribution (bell curve) in a given set of data.

### Negative skewed or left-skewed

```python
s = skewnorm.rvs(-5,loc=0,scale=1,size=1000)
# s
sns.kdeplot(s)
```

### Positive skewed or right-skewed  

```python
s = skewnorm.rvs(5,loc=0,scale=1,size=1000)
# s
sns.kdeplot(s)
```

# Kurtosis
- Kurtosis refers to the degree of presence of outliers in the distribution.

```python
a = np.random.randint(1,500,300)
# sns.kdeplot(a)
sns.kdeplot(a,bw_adjust=1,label="bw=1,high-kurtosis")
sns.kdeplot(a,bw_adjust=3,label="bw=3,low-kurtosis")
sns.kdeplot(a,bw_adjust=6,label="bw=6,low-kurtosis")
plt.legend()
plt.show()
```
