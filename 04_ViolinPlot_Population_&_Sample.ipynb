```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head(3)
```

```python
sns.boxplot(y=df.petal_length,width=.3)
```

```python
sns.kdeplot(df.petal_length)
```

# Violin Plot

```python
plt.subplot(121)
sns.violinplot(x=df.petal_length)
plt.subplot(122)
sns.violinplot(y=df.petal_length)
plt.show()
```

```python
plt.subplot(131)
sns.violinplot(x=df.petal_length,inner="quartile")
plt.subplot(132)
sns.violinplot(y=df.petal_length,inner="stick")
plt.subplot(133)
sns.violinplot(y=df.petal_length,inner="point")
plt.show()
```

# Population

```python
a = np.random.randint(1,100,1000)
plt.figure(figsize=(16,4))
plt.scatter(range(len(a)),a,alpha=.8)
plt.show()
```

```python
print("mean of Population",np.mean(a))
```

# Sample

```python
sample = np.random.choice(a,500)
plt.figure(figsize=(16,4))
plt.scatter(range(len(sample)),sample,alpha=.3)
plt.show()
```

```python
sample = np.random.choice(a,500)
plt.figure(figsize=(16,4))
plt.scatter(range(len(a)),a,alpha=.8,color="k",label="population")
plt.scatter(range(len(sample)),sample,alpha=.3,color="r",label="sample")
plt.legend()
plt.show()
```

```python
print("mean of Population",np.mean(sample))
```

# Mean of Population and Sample

```python
a = np.random.randint(1,100,1000)
sample = np.random.choice(a,500)
print("mean of Population",np.mean(a))
print("mean of Sample",np.mean(sample))
```
