# PDF & CDF of Gausian Normal Distribution

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# PDF

```python
a = np.random.normal(size = 100)
# 
```

```python
pdf = norm.pdf(a)
```

```python
sns.lineplot(x=a,y=pdf)
```

```python
sns.kdeplot(np.random.normal(size = 100),bw_adjust=1,label="$\mu=0,\sigma=1$")
sns.kdeplot(np.random.normal(size = 100),bw_adjust=2,label="$\mu=0,\sigma=2$")
sns.kdeplot(np.random.normal(size = 100),bw_adjust=3,label="$\mu=0,\sigma=3$")
plt.legend()
plt.show()
```

# CDF

```python
a = np.random.normal(size = 100)
pdf = norm.pdf(a)
cdf = norm.cdf(a)
sns.lineplot(x=a,y=pdf,label="pdf")
sns.lineplot(x=a,y=cdf,label="cdf")
plt.legend()
plt.show()
```
